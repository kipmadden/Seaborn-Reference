```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
dir(datasets)
```

```python
planets = (sns.load_dataset("planets").query("method in ['Radial Velocity', 'Transit']"))
```

```python
planets.head()
```

```python
ax = sns.histplot(planets, x="distance", hue="method", log_scale=True)
```

```python
ax.set(xlim=(.5,2e4),
      xlabel="Distance from Earth (pc)",
      yticks=[0,25,50,75,100, 125]);
```

```python
ax = sns.displot(planets, x="distance", log_scale=True, col="method")
```

```python
sns.set(color_codes=True)
f, ax = plt.subplots(figsize=(7, 7))
ax.set(xscale="log", yscale="log")
sns.regplot(y="orbital_period", x="mass", data=planets, ax=ax)
```

```python
df = pd.DataFrame(datasets.load_wine().data)
```

```python
df.columns = datasets.load_wine().feature_names
```

```python
df.head()
```

```python
fig, ax = plt.subplots()
ax.hist(df['alcohol'])
```

```python
df['alcohol'].plot.hist()
```

```python
sns.distplot(df['alcohol'], hist=False, rug=True, kde_kws={'shade':True} )
```

```python
df.columns
```

```python
print(df.apply(lambda col: len(col.unique())))
```

```python
for col in df:
    print(f'{len(df[col].unique())}\t: {col} ')
```

```python
sns.set(color_codes=True)
sns.lmplot(x="alcohol", y="ash", data=df)
```

```python
for style in ['white','dark','whitegrid','darkgrid','ticks']:
    sns.set_style(style)
    sns.displot(df['alcohol'], color='g')
    sns.despine(left=True)
    plt.show()
```

```python
for p in sns.palettes.SEABORN_PALETTES:
    sns.set_palette(p)
    sns.palplot(sns.color_palette())
    plt.show()
```

```python
sns.palplot(sns.color_palette("Blues", 12))
plt.show()
```

```python
fig, ax = plt.subplots()
sns.distplot(planets['orbital_period'], ax=ax)
ax.set(xscale="log", xlabel="mass")
plt.show()
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=1,
                              ncols=2,
                              sharey=True,
                               sharex=True,
                              figsize=(7,4))
sns.distplot(df['alcohol'], ax=ax0)
sns.distplot(df.query("11 <= alcohol <= 13")['alcohol'], ax=ax1)

ax1.set(xlabel="Alcohol (All)")
ax1.axhline(y=0.5, label="My limit", linestyle='--')
ax1.legend()
plt.show()
```

```python
# Create a plot with 1 row and 2 columns that share the y axis label
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, sharey=True)

# Plot the distribution of 1 bedroom apartments on ax0
sns.distplot(df['alcohol'], ax=ax0)
ax0.set(xlabel="Alcohol content")

# Plot the distribution of 2 bedroom apartments on ax1
sns.distplot(df['malic_acid'], ax=ax1, color='g')
ax1.set(xlabel="Malic Acid content")

# Display the plot

plt.show()
```
