```python
g = sns.JointGrid(data=df.query('clarity == "IF" and cut == "Ideal"'),
                 x="price",
                 y="carat")
g.plot(sns.regplot, sns.kdeplot)
```

```python
g = sns.JointGrid(data=df.query('clarity == "IF" and cut == "Ideal"'),
                 x="price",
                 y="carat")
g = g.plot_joint(sns.kdeplot)
g = g.plot_marginals(sns.kdeplot, shade=True)
#g = g.annotate(stats.pearsonr)
g.ax_joint.annotate(stats.pearsonr, xy=(0.1, 2))
plt.tight_layout()
```
